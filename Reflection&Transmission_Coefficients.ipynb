# Reflection & Transmission Coefficients

In this code I ilustrate how the reflection and transmission coefficients look for a potencial barrier with a particle with energy lower that the barrier.

For this, let's consider the expression:
$$T = \frac{ V_{0}^{2}\sinh^{2}(k_{1}a)  }{ 4E(V_{0}-E) + V_{0}^{2}\sinh^{2}(k_{1}a)  } $$
Where:
$$k_{1} = \sqrt{\frac{2m(V_{0}-E)}{\hbar^{2}}} $$
And the expression:
$$R + T = 1 $$

Let's take an electron $m_{e}$ and an barrier energy of $50 kJ$, the electron energy corresponds to $10 kJ$

```python
# Libraries used
import math
import numpy as np
import matplotlib.pyplot as plt


# Lets write the constant values:
m = 5
v0 = 10
E = 2
hbar = 1
a = 2

def k1 (m=m, v0=v0, E=E):
  k1 = math.sqrt( ( 2*m*(v0 - E)  )/(hbar**2)  )

  return k1


def transmission(v0=v0, k=k1(), a=a, E=E):
  T = ( (v0**2) * math.sinh(k*a)  ) / ( 4*E*(v0 - E) + v0**2 * (math.sinh(k*a))**2  )
  return T

def reflection(x):
  R = 1 - transmission(a=x)
  return R

domain = np.arange(0, a, 0.1)
t_values = np.array([ transmission(a=x) for x in domain  ])
r_values = np.array([ reflection(x) for x in domain  ])

plt.plot(domain, t_values, label = "Transmission values")
plt.plot(domain, r_values, label = "Reflection values")
plt.grid()
plt.xlabel("Barrier width $a$")
plt.ylabel("Transmission value T")
plt.legend()

plt.show()
```
